```python
import pandas as pd
import requests as rq
import json
from io import StringIO
from datetime import timedelta, date, datetime
import numpy as np
from sklearn import preprocessing
import plotly.graph_objects as go
```

```python
all_indices = pd.read_csv('https://eodhistoricaldata.com/api/exchanges/INDX?api_token=&fmt=csv')
print(all_indices)
```

```python
indices = all_indices#all_indices[all_indices['Country'] != 'Unknown']
print(indices)
```

```python
index_quotes = pd.DataFrame(columns=all_indices['Code'])
index_quotes['Date'] = None

index_quotes = index_quotes.set_index('Date')
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2019,12, 31)
end_date = datetime.now().date()
for single_date in daterange(start_date, end_date):
    day = single_date.strftime("%Y-%m-%d")
    print(day)
    daily_quotes = rq.get('http://eodhistoricaldata.com/api/eod-bulk-last-day/INDX?api_token=&fmt=csv&date='+day)
    df = pd.read_csv(StringIO(daily_quotes.text))
    df = df.set_index('Code')
    #print(df)
    quotes = []
    for code in index_quotes.columns:
        if code in df.index:
            quotes.append(df.loc[code]['Close'])
        else:
            quotes.append(None)
    index_quotes.loc[day] = quotes
    #print(index_quotes)
```

# Process Data

```python
new_index_quotes = pd.read_csv('indices-quotes.csv')
new_index_quotes.set_index('Date', inplace=True)
#print(new_index_quotes)
```

```python
cols = new_index_quotes.columns
for c in cols:
    quotes = new_index_quotes[c]
    #quotes = quotes.reset_index(drop=True)
    #print(quotes)
    #quotes = quotes.dropna()
    non_nan_count = np.count_nonzero(~np.isnan(quotes))
    if non_nan_count < 20:
        del new_index_quotes[c]
#print(new_stock_quotes)     
```

```python
new_index_quotes = new_index_quotes.fillna(method='ffill')
print(new_index_quotes)
```

```python
start_date = '2020-02-14'
today = datetime.now().date().strftime("%Y-%m-%d")

new_index_quotes = new_index_quotes.loc[start_date:today]
normalized = new_index_quotes.apply(lambda x: (x / x.iloc[1]))
#print(normalized.loc['2020-03-05'])
#normalized = normalized.drop(['2020-03-05'])
```

# By Country

```python
def avg_symbols_perf(symbols, data):
    #print(symbols)
    #print(data)
    df = pd.DataFrame(columns=symbols)
    df2 = data.filter(items=df.columns)#pd.merge(df,data, on=df.columns, how="inner")
    #print(df2)
    #print(df2.loc['2020-03-05'])
    df2 = df2.mean(axis=1)
    df2 = df2.replace([np.inf, -np.inf], np.nan)
    #df2 = df2.dropna()
    print(df2)
    return df2
```

```python
by_country = {}
for x,i in indices.iterrows():
    country = i["Country"]
    if country not in by_country:
        by_country[country] = []
    by_country[country].append(i["Code"])

#print(by_country)
```

```python
returns_by_country = pd.DataFrame(columns=by_country.keys())
for k in by_country:
    print(k)
    print(by_country[k])
    
    returns_by_country[k] = avg_symbols_perf(by_country[k],normalized)
    

#returns_by_sector['Financial Services'] = None
#returns_by_sector['Basic Materials'] = None

print(returns_by_country)
```

```python
fig = go.Figure()
#print(returns_by_sector)
for c in returns_by_country.columns:
    #print(c)
   # print(returns_by_sector[k])
    fig.add_trace(go.Scatter(
                    x=returns_by_country.index,
                    y=returns_by_country[c],
                    name=c,
                    #line_color='deepskyblue',
                    opacity=1))
fig.show()
```
